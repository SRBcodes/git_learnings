```python
import pandas as pds
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
url = 'https://raw.githubusercontent.com/sai1krishna/Assignment/main/Caravan.csv'
csv = pds.read_csv(url,index_col=0,parse_dates=[0])
```

```python
csv=csv.replace(to_replace="No",value="0")
csv=csv.replace(to_replace="Yes",value="1")
csv
```

```python
y_values = csv['Purchase']
x_values = csv.drop(['Purchase'],axis=1)
x_training,x_test,y_training,y_test = train_test_split(x_values, y_values, test_size = 0.2 , random_state = 1)
csv['Purchase'].value_counts()
```

```python
Logistic_cls = LogisticRegression(solver='liblinear')
Logistic_cls.fit(x_training, y_training)
```

```python
x1_values = Logistic_cls.predict(x_test)
print(x1_values)
print([ a for a in x1_values if a == 1])
```

```python
Logistic_cls.score(x_test,y_test)
```
